```python
from sentence_transformers import SentenceTransformer

encoder = SentenceTransformer("all-mpnet-base-v2")
vectors = encoder.encode(df.text)
vectors.shape
```

```python
vectors
```

```python
dim = vectors.shape[1]
dim
```

```python
import faiss

index = faiss.IndexFlatL2(dim)
index
```

```python
index.add(vectors)
```

```python
search_query="I want to buy a polo t-shirt"

vec = encoder.encode(search_query)
vec.shape
```

```python
import numpy as np
svec = np.array(vec).reshape(1,-1)
svec.shape
```

```python
distances, I = index.search(svec, k=2)
I
```

```python
import os
import streamlit as st
import pickle
import time
import langchain
from langchain.llms import GooglePalm
from langchain.chains import RetrievalQAWithSourcesChain
from langchain.chains.qa_with_sources.loading import load_qa_with_sources_chain
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import SeleniumURLLoader
from langchain.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceHubEmbeddings
from langchain.chains import RetrievalQA
```

```python
os.environ['GOOGLE_API_KEY'] = "enter your secret api key here"
```

```python
llm = GooglePalm(temperature=0.9, max_tokens=500)

loaders = SeleniumURLLoader(urls=[
    "https://www.moneycontrol.com/news/business/markets/wall-street-rises-as-tesla-soars-on-ai-optimism-11351111.html",
    "https://www.moneycontrol.com/news/business/tata-motors-launches-punch-icng-price-starts-at-rs-7-1-lakh-11098751.html",
])
data = loaders.load()
len(data)
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)

docs = text_splitter.split_documents(data)
len(docs)
```

```python
docs[0]
```

```python
docs[9]
```

```python
docs[30]
```

```python
from huggingface_hub import notebook_login

notebook_login()
```

```python
embeddings = HuggingFaceHubEmbeddings()

vectorindex = FAISS.from_documents(docs, embeddings)
vectorindex
```

```python
file_path="vector_index.pkl"
with open(file_path, "wb") as f:
    pickle.dump(vectorindex, f)
```

```python
if os.path.exists(file_path):
    with open(file_path, "rb") as f:
        vectorIndex = pickle.load(f)
```

```python
retriever = vectorIndex.as_retriever()
chain = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=retriever, input_key="query", return_source_documents=True)

query = "What is the price of Tiago Punch?"
langchain.debug=True
chain(query)
```
